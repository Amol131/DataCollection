## Extract information from the given web site

You will extract the data from the below web site: <br>


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

Download the webpage at the url


```python
#your code goes here
print(requests.get(url))          #this just gives us the response_code. 200means code executed successfully.
data = requests.get(url).text
print(data)
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data, 'html5lib')
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
table = soup.find('table')
mini_lis = []
lis = []
for row in table.find_all('tr'):
    cols = row.find_all('td')
    Language_name = cols[1].getText()
    annual_average_salary = cols[3].getText()
    mini_lis = []
    mini_lis.append(Language_name)
    mini_lis.append(annual_average_salary)
    lis.append(mini_lis)
    print(f'{Language_name} >> {annual_average_salary}')
print(lis)

# Here since our aim is to get data in dataframe format, first we converted available data into list of lists form... and then converted it to DataFrame. 
# Our lis variable contains the data in the required form. 
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
df = pd.DataFrame(lis)      #converted into dataframe
df.columns = df.iloc[0]     #make the first row header 
df = df[1:]                 #Removing unnecessary first row
print(df)
df.to_csv("popular-languages.csv")

```
